```python
AGV = np.array(df["Agriculture Value"])

print("Total: ", np.sum(AGV)) #Sum
print("Mean: ", np.mean(AGV))#mean
print("Maximum: ", np.max(AGV))#maximum
print("Minimum: ", np.min(AGV))#Minimum
print("Standard Deviation: ", np.std(AGV))#Standard deviation
```

```python
OilRev = np.array(df["Oil Rents"])

print("Total: ", np.sum(OilRev)) #Sum
print("Mean: ", np.mean(OilRev))#mean
print("Maximum: ", np.max(OilRev))#maximum
print("Minimum: ", np.min(OilRev))#Minimum
print("Standard Deviation: ", np.std(OilRev))#Standard deviation
```

```python
# Select variables
stats = df[["GDP growth",
            "Agriculture Value",
            "Oil Rents",
            "Inflation Rate",
            "Exchange Rate",
           "GDP (US$)"]]

# Generate descriptive statistics
summary_table = pd.DataFrame({
    'Mean': stats.mean(),
    'Minimum': stats.min(),
    'Maximum': stats.max(),
    'Std Dev': stats.std()
})

print(summary_table)
```

```python

fig, ax = plt.subplots(figsize=(10,4))
ax.axis('off')

table = ax.table(
    cellText=summary_table.round(2).values,
    colLabels=summary_table.columns,
    rowLabels=summary_table.index,
    loc='center'
)

table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.5)

plt.savefig("descriptive_statistics.png",
            bbox_inches='tight',
            dpi=300)

plt.show()
```
